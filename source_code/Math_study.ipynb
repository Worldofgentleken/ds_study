# 변수와 상수

```python
import numpy as np
```

```python
a = 1.5  # a : 상수
x = np.linspace(-1, 1) # x : 변수 -1 부터 1의 범위
y = a * x # : y : 변수

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

# 함수
- 함수는 수식을 다루는데 기본이 되는 개념이다.
- 함수는 x를 정하면 그것에 종속적인 값 y가 정해지는 관계를 말한다.
- 프로그래밍에서의 함수에는 함수에 들어가는 값으로서 인수가 있고, 함수에서 나오는 값으로서 반환값이 있다.

```python
def my_func(x):
    return 3*x + 2

x = 4
y = my_func(x)
print(y)
```

# 거듭제곱과 제곱근
- 같은 수 또는 문자를 여러 번 곱하는 것을 거듭제곱 이라고 한다.
- 2제곱해서 x가 되는 y를 x의 제곱근 이라고 한다.

```python
def my_func(x):
    a = 3
    return x**a

x = np.linspace(0, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return np.sqrt(x)

x = np.linspace(0, 9)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return np.sqrt(x) + 1

x = np.linspace(0, 9)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

# 다항식 함수

```python
def my_func(x):
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    a = 4*x**3 + 2*x**2 + x + 3
    return a

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
def my_func(x):
    a = x**3 - 2*x**2 + - 3*x + 4
    return a

x = np.linspace(-2, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.show()
```

# 삼각함수

```python
def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi)
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label = "sin")
plt.plot(x, y_cos, label = "cos")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.show()
```

```python
def my_tan(x):
    return np.tan(x)

x = np.linspace(-1.3, 1.3)
y_tan = my_tan(x)

plt.plot(x, y_tan, label = "tan")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.show()
```

# 총합과 총곱

```python
a = np.array([1, 2, 3, 4, 5])
y = np.sum(a)
print(y)
```

```python
a = np.array([1, 2, 3, 4, 5])
y = np.prod(a)
print(y)
```

```python
b = np.array([1, 2, 3, 4, 5, 6])
y_sum = np.sum(b)
y_prod = np.prod(b)
print(y_sum)
print(y_prod)
```
